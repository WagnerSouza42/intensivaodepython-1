```python
import pyautogui
import pyperclip
import time

pyautogui.PAUSE = 1

# Passo 1: Entrar no sistema (no nosso caso, entrar no link)
pyautogui.hotkey("ctrl", "t")
pyperclip.copy("https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing")
pyautogui.hotkey("ctrl", "v")
pyautogui.press("enter")

time.sleep(5)

# Passo 2: Navegar até o local do relatório (entrar na pasta Exportar)

pyautogui.click(x=342, y=274, clicks=2) # Exportar
time.sleep(2)

# Passo 3: Fazer o download do relatório
pyautogui.click(x=379, y=358)#documento base exel
pyautogui.click(x=1159, y=169)# posição dos 3 pontinho  
pyautogui.click(x=952, y=563)# fazer dowload
time.sleep(5)

# com esse codigo eu eu posso acessar a base de dados para baixar e e fazer analise.
```

```python
# Passo 4: Calcular os indicadores
import pandas as pd

tabela = pd.read_excel(r"C:\Users\vini_\Downloads\Vendas - Dez.xlsx")#colocar Aqui o caminho onde base de dados foi salva.
display(tabela) # usei um display tabela para visualizar minha tabela e fazer o calculo dos indicadores 

faturamento = tabela["Valor Final"].sum() # aqui mostrei qual coluna meu codigo usaria
quantidade = tabela["Quantidade"].sum()
print(quantidade)
print(faturamento)
```
